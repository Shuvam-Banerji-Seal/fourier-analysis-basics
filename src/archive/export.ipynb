```python
# Imports:
import os
import nbformat
from nbconvert import PDFExporter
from traitlets.config import Config

# Get versuch_nr:

# Export file data:
file_path = "../presentation/notebook.pdf"
title = f"Jupyter-Notebook zu Fourier und SageMath"

# Delete current file:
try:
    os.remove(file_path)
except:
    pass

# Configuration
custom_meta = {"name": title}
c = Config()
# c.PDFExporter.exclude_input = True
# see https://stackoverflow.com/q/70949216/9655481
c.LatexPreprocessor.date = ""

# Do pdf export:
exporter = PDFExporter(config=c)
with open("fourier.ipynb", "r") as infile:
    nb = nbformat.reads(infile.read(), nbformat.NO_CONVERT)
(source, meta) = exporter.from_notebook_node(nb, resources={"metadata": custom_meta})
with open(file_path, "wb") as f:
    f.write(source)
    f.close()
```
